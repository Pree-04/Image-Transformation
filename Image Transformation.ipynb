# Image Transformation

## Applying Geometrical Transformations On An Image.

# Translation

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('king kong.jpg')
rows,cols,ch = img.shape

M = np.float32([[1,0,200],[0,1,200]])
dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('After Translation')
plt.show()
```

## Scaling

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('king kong.jpg')


height, width = img.shape[:2]
print(img.shape[:2])

#Scaled down
res = cv2.resize(img,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)

#scale up
#res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)


plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(res),plt.title('After scaling')
plt.show()
# Notice that the X and Y values are scalled up
```

# Rotation

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('king kong.jpg')
rows,cols,ch = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1) #getRotationMatrix2D(center,angle,scale = 1 ).

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('After Rotation')
plt.show()
```
